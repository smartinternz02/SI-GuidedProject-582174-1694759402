## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = dataset[['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction', 'EnvironmentSatisfaction','YearsWithCurrManager','WorkLifeBalance']]
Y = dataset['Attrition']
```

```python
X.head()
```

```python
Y.head()
```

## Splitting data into test and train

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=None)
```

```python
X_train,X_test,Y_train,Y_test.shape
```

# Logistic Regression

## Model Building & Import the model building Libraries

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

model.fit(X_train, Y_train)
```

```python
pred=model.predict(X_test)
pred[:5]
```

### Evaluation of classification model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
accuracy = accuracy_score(Y_test, pred)
report = classification_report(Y_test, pred, zero_division=1)
print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')
```

```python
confusion_matrix(Y_test,pred)
```

```python
pd.crosstab(Y_test,pred)
```

## Roc-AUC curve

```python
probability=model.predict_proba(X_test)[:,1]
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
Y_test_bin = lb.fit_transform(Y_test)
fpr, tpr, thresholds = roc_curve(Y_test_bin, probability)


plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

dt_model = DecisionTreeClassifier(random_state=50,max_depth=4,min_samples_split=50)

dt_model.fit(X_train, Y_train)
```

```python
dt_predictions = dt_model.predict(X_test)

dt_accuracy = accuracy_score(Y_test, dt_predictions)

dt_report = classification_report(Y_test, dt_predictions)

print(f'Decision Tree Accuracy: {dt_accuracy}')
```

```python
print(f'Decision Tree Classification Report:\n{dt_report}')
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(max_features=12)
rf_model.fit(X_train, Y_train)
```

```python
rf_predictions = rf_model.predict(X_test)
rf_accuracy = accuracy_score(Y_test, rf_predictions)
rf_report = classification_report(Y_test, rf_predictions)
print(f'Random Forest Accuracy: {rf_accuracy}')
```

```python
print(f'Random Forest Classification Report:\n{rf_report}')
```
