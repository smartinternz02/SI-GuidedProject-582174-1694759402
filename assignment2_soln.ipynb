#**Univariate Analysis**

checking for box plot in all columns to check for any outliers

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(12)

plt.subplot(3,3,1)
sns.boxplot(y="total",data=dataset)

plt.subplot(3,3,2)
sns.boxplot(y="speeding",data=dataset)

plt.subplot(3,3,3)
sns.boxplot(y="alcohol",data=dataset)

plt.subplot(3,3,4)
sns.boxplot(y="no_previous",data=dataset)

plt.subplot(3,3,5)
sns.boxplot(y="not_distracted",data=dataset)

plt.subplot(3,3,6)
sns.boxplot(y="ins_premium",data=dataset)

plt.subplot(3,3,7)
sns.boxplot(y="ins_losses",data=dataset)
```

inference:

there are 3 outliers in alcohol column and 1 outlier in not_distracted column

histogram analysis

```python
f = plt.figure()
f.set_figwidth(16)
f.set_figheight(11)

plt.subplot(3,3,1)
plt.hist(x="alcohol",data=dataset)

plt.subplot(3,3,2)
plt.hist(x="total",data=dataset)

plt.subplot(3,3,3)
plt.hist(x="ins_premium",data=dataset)

plt.subplot(3,3,4)
plt.hist(x="ins_losses",data=dataset)

plt.subplot(3,3,5)
plt.hist(x="not_distracted",data=dataset)

plt.subplot(3,3,6)
plt.hist(x="no_previous",data=dataset)

plt.subplot(3,3,7)
plt.hist(x="speeding",data=dataset)
```

**inference:**

the values of alcohol and ins_premium are more varied

where as the remainig values are little well-spread

now we check the nature of the data.

```python
j = 1
plt.figure(figsize=(15,10))
for i in dataset.drop(columns=['abbrev']):
    plt.subplot(3,3,j)
    plt.xlabel(i)
    sns.kdeplot(data = dataset[i])
    j += 1
plt.show()
```

**inference:**
the data exhibits a nearly normal distribution like nature

#**Multivariate analysis**

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Create barplots for each column
sns.barplot(data=dataset, x='speeding', y='total', ax=axes[0, 0])
axes[0, 0].set_title('Speeding vs Total')

sns.barplot(data=dataset, x='alcohol', y='total', ax=axes[0, 1])
axes[0, 1].set_title('Alcohol vs Total')

sns.barplot(data=dataset, x='not_distracted', y='total', ax=axes[1, 0])
axes[1, 0].set_title('Not Distracted vs Total')

sns.barplot(data=dataset, x='ins_premium', y='total', ax=axes[1, 1])
axes[1, 1].set_title('Premium vs Total')
```

**inference:**

alcohol and not_distracted can affect the car crashes

but it seems that insurance premium and speeding doesn't show much affect

```python
f = plt.figure()
f.set_figwidth(21)
f.set_figheight(14)
sns.barplot(data=dataset,x="total",y="alcohol",hue="speeding")
```

inference:

alcohol and has significant affect on total

speeding doesn't have significant affect on total

Again, both alcohol and speeding together have a significant affect on total car crashes
